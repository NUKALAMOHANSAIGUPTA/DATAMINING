```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/customer_purchases.csv'
df = pd.read_csv(file_path)
```

```python
print("Dataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Annual_Income'], bins=30, kde=True, color='blue')
plt.title('Histogram of Annual Income')
plt.xlabel('Annual Income (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Purchase_Frequency'], y=df['Spending_Score'], hue=df['Gender'])
plt.title('Purchase Frequency vs. Spending Score')
plt.xlabel('Purchase Frequency (per month)')
plt.ylabel('Spending Score')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
corr_matrix = df[['Annual_Income', 'Spending_Score', 'Purchase_Frequency']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
outliers = df[df['Spending_Score'] > df['Spending_Score'].quantile(0.95)]
print("\nOutliers (Top 5% Spending Score):")
print(outliers)
```
